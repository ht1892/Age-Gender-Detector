```python
import cv2 as cv
from matplotlib import pyplot as plt
```

PLOTTING

```python
img=cv.imread("Data/a2.jpg")
```

```python
plt.imshow(img)
```

```python
img2=cv.imread("Data/k.jpg")[0:1280,400:1120]
plt.imshow(img2)
```

Arithmatic Operation

```python
#Addition
result=img+img2
plt.imshow(result)
```

```python
#Substraction
result=img-img2
plt.imshow(result)
```

```python
#Multiplication
plt.imshow(img*img2)
```

```python
#Division
plt.imshow(img/img2)
```

BLURRING

```python
img=cv.cvtColor(img,cv.COLOR_BGR2HSV)
figure_size=7
result=cv.blur(img,(figure_size,figure_size))
plt.figure(figsize=(11,6))
plt.subplot(122),plt.imshow(cv.cvtColor(img,cv.COLOR_HSV2BGR)),plt.title("Original")
plt.xticks([]),plt.yticks([])
plt.subplot(131),plt.imshow(cv.cvtColor(img,cv.COLOR_HSV2BGR)),plt.title("Blurred")
plt.xticks([]),plt.yticks([])
plt.show()
```

ZOOM OR SCALING

```python
(height,width)=img.shape[0:2]
zoom=cv.resize(img,None,fx=1.5,fy=1.5,interpolation=cv.INTER_CUBIC)
plt.imshow(zoom)
```

RESIZING

```python
result=cv.resize(img,(int(width*2),int(height*2)),interpolation=cv.INTER_CUBIC)
plt.imshow(result)
```

ROTATION

```python
m=cv.getRotationMatrix2D((width/2,height/2),30,1)
rotate=cv.warpAffine(img,m,(height,width))
plt.imshow(rotate)
```

TRANSLATE IMAGE OR MOVING TO CERTAIN POINT
